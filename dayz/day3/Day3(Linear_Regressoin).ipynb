```python
import csv
```

```python
import matplotlib.pyplot as plt
```

```python
hours_study = []
marks = []
with open('dataset/regression_training.csv',mode ='r')as file:
    csvFile = csv.reader(file)
    next(csvFile)
    for lines in csvFile:
        hours_study.append(lines[0])
        marks.append(lines[-1])
```

```python
plt.scatter(hours_study,marks)
plt.show()
```

```python
n = len(hours_study)

X = [int(x) for x in hours_study]
Y = [int(y) for y in marks]

x_mean = sum(X) / n
y_mean = sum(Y) / n        
numerator = 0
denominator = 0

for i in range(n):
    numerator += (X[i] - x_mean) * (Y[i] - y_mean)
    denominator += (X[i] - x_mean) ** 2

if denominator == 0:
    m = 0
else:
    m = numerator / denominator
    
c = y_mean - m * x_mean

print(m,c)
```

```python
hours_study_test = []
with open('dataset/regression_test.csv',mode ='r')as file:
    csvFile = csv.reader(file)
    next(csvFile)
    for lines in csvFile:
        hours_study_test.append(lines[0])

X_test = [int(x) for x in hours_study_test]
```

```python
final_rez = []
for x in X_test:
    rez = m * x + c
    final_rez.append(rez)

print(final_rez)
```

```python
plt.scatter(X,Y,color="violet",label='Train')
plt.scatter(X_test,final_rez,color='purple',label="Test")
plt.axline((0, c), slope=m,)
plt.show()
```
